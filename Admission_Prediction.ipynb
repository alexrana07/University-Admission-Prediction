Pandas and NumPy are used for handling and processing data.

# DATA CLEANING

```python
#Check Missing Values
df.isnull().sum()
```

Result: No missing values (dataset is clean)

```python
#Rename Columns
df.rename(columns={
    'GRE Score': 'GRE',
    'TOEFL Score': 'TOEFL',
    'University Rating': 'University_Rating',
    'LOR ': 'LOR',
    'Chance of Admit ': 'Chance_of_Admit'
}, inplace=True)
```

We rename the column names to make them clean and consistent so Linear Regression can use them easily.
Some original column names contain spaces and extra characters, which can cause confusion or errors while selecting features and interpreting results.

```python
#Remove Unnecessary Columns
df.drop(columns=['Serial No.'], inplace=True)
```

we remove the Serial Number column because it does not contain any useful information for predicting admission chances.

# FEATURE SELECTION :

```python
X = df.drop('Chance_of_Admit', axis=1)
y = df['Chance_of_Admit']
```

Feature selection is the process of choosing which columns (features) from the dataset will be used to predict admission.  
X (input features): all the student details like GRE, TOEFL, CGPA, etc.    
y (target feature): the Chance of Admit that we want the model to predict.  
This separation is necessary so the model knows what to learn from (X) and what to predict (y).

# FEATURE SCALING :

Linear Regression works better with scaled data.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Feature scaling standardizes all input features (GRE, TOEFL, CGPA, etc.) so that they are on the same scale.     
This prevents features with larger numbers from dominating the model.  
StandardScaler transforms the data so most values are centered around 0 with a similar range.  
Scaling improves the accuracy and stability of the model.

# TRAIN-TEST SPLIT :

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

we divide the dataset into two parts:

Training data: used to teach the model (80% of the data)  
Testing data: used to check how well the model predicts new, unseen data (20% of the data)  
This ensures that the model can learn from some data and be tested on data it hasnt seen, which helps us evaluate its real performance.

# MODEL BUILDING  LINEAR REGRESSION :

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

we create a Linear Regression model and train it using the training data.  
The model learns how each feature (GRE, CGPA, TOEFL, etc.) affects the Chance of Admit.  
Essentially, it finds the best-fitting line that predicts admission chances based on student data.  
Once trained, the model can make predictions on new student profiles.

# MODEL PREDICTION :

```python
y_pred = model.predict(X_test)
```

we use the trained model to predict the admission chances for the test data.  
The model looks at the student details in X_test and estimates the Chance_of_Admit.  
These predictions are later compared with the actual values to see how accurate the model is.

# MODEL EVALUATION :

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

we check how well the models predictions match the actual admission chances.  
MAE (Mean Absolute Error): average difference between predicted and actual values.  
MSE (Mean Squared Error): average squared difference; larger errors are penalized more.  
RMSE (Root Mean Squared Error): the square root of MSE; gives error in the same units as the target.  
R Score: shows how much of the variation in admission chances the model can explain (closer to 1 is better).  
These metrics tell us whether the model is accurate and reliable.

# ACTUAL vs PREDICTED VISUALIZATION :

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Admission Chance")
plt.ylabel("Predicted Admission Chance")
plt.title("Actual vs Predicted Admission Chances")
plt.show()
```

This scatter plot compares the actual admission chances (from the test data) with the predicted chances from the model.  
Each dot represents one student.    
If the model is perfect, all dots would lie on a straight diagonal line.  
The closer the dots are to the diagonal, the more accurate the model is.

# FEATURE IMPORTANCE (INTERPRETATION) :

```python
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})

feature_importance.sort_values(by='Coefficient', ascending=False)
```

we check how much each feature affects the admission chances.  
The coefficients from the linear regression model tell us the impact of each feature.  
A higher positive coefficient means that the feature increases the chance of admission.  
A negative coefficient would mean the feature reduces admission chances.  
This helps us understand which student details are most important for predicting admission.

```python
new_student = np.array([[320, 110, 4, 4.5, 4.5, 9.2, 1]])
new_student_scaled = scaler.transform(new_student)

prediction = model.predict(new_student_scaled)
print("Predicted Chance of Admit:", prediction[0])
```
